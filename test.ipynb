```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gpxpy
import geopandas as gpd
import folium
```

```python
gpx_path = 'route_2022-02-08_6.29pm.gpx'
with open(gpx_path) as f:
    gpx = gpxpy.parse(f)

```

```python
points = []
for segment in gpx.tracks[0].segments:
    for p in segment.points:
        points.append({
            'time': p.time,
            'latitude': p.latitude,
            'longitude': p.longitude,
            'elevation': p.elevation,
        })
df = pd.DataFrame.from_records(points)

gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))

gdf
```

```python
m = folium.Map(location=[df.latitude.mean(), df.longitude.mean()], zoom_start=15)

# Create a list of coordinates
coordinates = list(zip(df.latitude, df.longitude))

#Create a marker
folium.Marker([43.635457, 7.137412], popup='MARINA BAIE DES ANGES').add_to(m)

# Add the path to the map
folium.PolyLine(coordinates, color='blue').add_to(m)

m
```
